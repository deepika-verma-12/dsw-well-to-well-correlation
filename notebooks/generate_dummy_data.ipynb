```python
import os
import numpy as np
import pandas as pd

# 1. Get the exact folder path where this notebook is currently sitting
notebook_dir = os.getcwd()
print(f"Current folder path: {notebook_dir}")

files_to_create = [
    'All-NO 159-1 .xlsx', 'All-NO159-2.xlsx', 'All-J4ST2.xlsx', 
    'All-J6A.xlsx', 'All-J2ST1.xlsx', 'All_GS_172-2.xlsx', 
    'All-J5 .xlsx', 'All-J7A.xlsx'
]

print("\nGenerating safe synthetic datasets directly in this folder...")

for file in files_to_create:
    rows = 150 
    
    # Generate mock log curves matching your notebook's column index calls (Depth=0, GR=1, NPHI=2)
    mock_data = {
        'Depth': np.linspace(5000, 5150, rows),
        'GR (API)': np.random.uniform(20, 120, rows),       
        'Neutron Porosity': np.random.uniform(0.05, 0.40, rows)   
    }
    
    df = pd.DataFrame(mock_data)
    
    # 2. Force it to save exactly inside this directory
    full_save_path = os.path.join(notebook_dir, file)
    df.to_excel(full_save_path, index=False)
    print(f"-> Saved at: {full_save_path}")

print("\nInitialization Complete! The dummy data files are now visible in your workspace.")
```
